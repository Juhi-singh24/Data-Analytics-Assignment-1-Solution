Q 1. Why are dictionary keys required to be immutable in python?Explain with a small example.

Ans- Dictionary keys must not change, because Python uses the key’s hash value to store and find data.
If a key changes, Python won’t be able to find the value again.

Immutable means: value that cannot be changed (like int, str, tuple).

** Example **

```python
## Program 
d = {(1, 2): "Hello"}   # tuple is immutable

print(d[(1, 2)])
```

Q 2 . Write a program that takes a list of elements and create a dictionary showing the frequency of each element.


```python
## Example list
items = [1, 2, 2, 3, 1, 1]
freq = {}

for item in items:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1

print(freq)

```

 Q 3 What is the difference between dict.get(key) and dict[key]? Write a code example where get() is safer.

 Ans - 
      
    dict[key]:-

       * Directly accesses the value
       * Throws an error if the key does not exist

   **dict.get(key)**:-
   
      * Safely accesses the value
      * Returns None if key does not exist (or a default value)


**Code example where get( ) is safer**

```python

marks = {"Math": 85, "English": 90}

print(marks.get("Science", 0))
```

No error, program runs safely.

**Using dict[key] (unsafe)**


```python
marks = {"Math": 85, "English": 90}

print(marks["Science"])
```

Program crashes because "Science" key does not exist.

Q 4. Write a program to merge two dictionaries and resolve common keys by keeping the larger value.

```python
## Program
d1 = {"a": 10, "b": 20}
d2 = {"b": 25, "c": 15}
result = d1.copy()

for key, value in d2.items():
    if key in result:
        result[key] = max(result[key], value)
    else:
        result[key] = value
print("Merged Dictionary:")
print(result)
```

Q 5.Using a dictionary comprehension create a dictionary from a list of integers where keys are numbers and values are their cubes,but only for odd numbers.

```python
## Program:-
numbers = [1, 2, 3, 4, 5]
cube_dict = {x: x**3 for x in numbers if x % 2 != 0}

print("Cubes of odd numbers :-")
print(cube_dict)
```
